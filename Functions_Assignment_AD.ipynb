6. Explain the concept of generators in Python and how they are defined.


- A generator is a special function that produces values one at a time using the yield keyword. They don't store all values in memory but generate them on demand.

Example:

def counter():
    for i in range(3):
        yield i

gen = counter()

print(next(gen))  
- 0

print(next(gen))  
- 1

- Memory Efficient:
They generate values one by one — no need to store large lists.

- Faster for large data:
Useful for big files, streams.

- Lazy evaluation:
Values are produced only when needed.

Example:

    for i in range(5):

1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even_numbers([1,2,3,4,5,6]))
```

4) Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for num in range(1,201):
    if is_prime(num):
        print(num, end=" ")
```

5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, limit):
        self.limit = limit
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

for x in FibonacciIterator(10):
    print(x, end=" ")
```

6) Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(power_of_two(5)))
```

7) Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_generator(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

for l in read_file_generator("sample.txt"):
    print(l)
```

10) Create a Python program that uses filter() to remove all vowels from a string.

```python
text = "Hello World"
vowels = "aeiouAEIOU"

result = ''.join(filter(lambda x: x not in vowels, text))

print(result)
```

11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10 € if the value of the
order is smaller than 100 €.

Write a Python program using lambda and map.

```python
orders = [(34587, 40.95, 4),
          (98762, 56.80, 5),
          (77226, 32.95, 3),
          (88112, 24.99, 3)]

result = list(map(lambda x: (x[0], (x[1] * x[2]) if (x[1] * x[2]) > 100 else (x[1] * x[2]) + 10), orders))

print(result)
```
