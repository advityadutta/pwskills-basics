1. What is NumPy, and why is it widely used in Python?

- NumPy is a Python library used for numerical and mathematical operations.

- It provides support for arrays, matrices, and many mathematical functions.

- It is widely used because it is fast, memory-efficient, and allows vectorized operations, which makes data processing much quicker compared to normal Python lists.

----


2. How does broadcasting work in NumPy?


- Broadcasting allows NumPy to perform operations on arrays of different shapes. Instead of copying data, NumPy automatically expands the smaller array to match the larger one, this makes operations faster and avoids unnecessary memory usage.

---

3. What is a Pandas DataFrame?

- A Pandas DataFrame is a 2-dimensional, tabular data structure with rows and columns.

- It is similar to an Excel sheet or SQL table.

- Each column can store different data types like numbers, strings, or dates.

---

4. Explain the use of the groupby() method in Pandas?

- The groupby() method is used to group data based on one or more columns and then apply aggregation functions like sum(), mean(), or count().

- It is commonly used for data analysis and summarization.

6. What are the differences between NumPy arrays and Python lists?

- NumPy arrays store homogeneous data (same data type) while Python lists can store mixed data types

- NumPy arrays are faster and memory efficient while Lists are slower for numerical computations

# Practical Questions

1. How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

2. Write a Pandas script to find the mean of a specific column in a DataFrame.

```python
import pandas as pd

df = pd.DataFrame({
    'marks': [70, 80, 90, 85]
})

print(df['marks'].mean())
```

3. Create a scatter plot using Matplotlib?

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

corr = df.corr()
sns.heatmap(corr, annot=True)
```

5. Generate a bar plot using Plotly?

```python
import plotly.express as px

data = {'Category': ['A', 'B', 'C'],
        'Values': [10, 20, 30]}

fig = px.bar(data, x='Category', y='Values')
fig.show()
```

6. Create a DataFrame and add a new column based on an existing column

```python
import pandas as pd

df = pd.DataFrame({'salary': [30000, 40000, 50000]})
df['tax'] = df['salary'] * 0.1

print(df)
```

7. Write a program to perform element-wise multiplication of two NumPy arrays

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

11. Perform matrix multiplication using NumPy

```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

result = np.dot(a, b)
print(result)
```

13. Create a 3D scatter plot using Plotly

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z')
fig.show()
```
