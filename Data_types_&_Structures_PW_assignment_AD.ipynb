**Data Types and Structure Theoretical Questions**

1. What are data structures, and why are they important?
- Data Structures are organized way of storing and managing data so it can be accessed and used efficiently. they are important to make programs faster and handle large amount of data easily.

2. Explain the difference between mutable and immutable data types with examples.
- Mutable: can be changed after creation. example: lists in python
-Immutable: cannot be changed after creation. example: Tuples in python

3. What are the main differences between lists and tuples in Python?
-Lists are mutable, tuples are immutable
-Lists use[] while tuples use ()
- Lists are used for dynamic data while tuples for fixed data
-lists are slower while tuples are faster

4.Describe how dictionaries store data.
- Dictionaries stores data as key-value pairs.

5. Why might you use a set instead of a list in Python?
- Sets automatically remove duplicates
- Sets offer faster membership checks (in operations)
- sets are good when order does not matter.

6. What is a string in Python, and how is it different from a list?
- String is a sequence of characters and is immutable
- List is a collection of items and is mutable

7. How do tuples ensure data integrity in Python?
- Tuples are immutable, which prevents accidental modification and ensures consistent data.

8. What is a hash table, and how does it relate to dictionaries in Python?
- Hash table is a data structure that maps keys to values using a hash function. Python distionaries are implmented using hash tables, enabling fast lookups.

9. Can lists contain different data types in Python?
- Yes

10. Explain why strings are immutable in Python?
- Strings are immutable for performance and security reasons, Changing a string would require creating a new memory space, so Python keeps them fixed.

11. What advantages do dictionaries offer over lists for certain tasks?
- Faster searching using keys
- More meaningful data representation
- Good for lookups, indexing by name instead of by number

12. Describe a scenario where using a tuple would be preferable over a list.
- Tuples would be preferable when storing a data that should not change. example: Date of Birth for employees, Employee ID

13. How do sets handle duplicate values in Python?
- Sets automatically remove duplicates, storing only unique values.

14. How does the “in” keyword work differently for lists and dictionaries?
- in List, in checks values.
- in Dictionaries, in check keys, not values.

15. Can you modify the elements of a tuple? Explain why or why not?
- No, Tuples are immutable.

16. What is a nested dictionary, and give an example of its use case?
- Nested dictionary is a dictionary inside another dictionary.
- example: student = { "name": "Adv", "marks": {"maths": 90, "Science": 86}}

17. Describe the time complexity of accessing elements in a dictionary.
- Average time complexity is O(1) (constant time) because dictionaries use hash tables. Accessing elements in a dictionary has an average time complexity of O(1), also called constant time.

18. In what situations are lists preferred over dictionaries?
- when ordering matters
- when storing simple sequences
- when only index based access is required

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries are unordered because their internal structure (hash table) doesn't store items in a fixed order. This means you cannot rely on insertion order for logical accessing (even though newer versions maintain order).

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- list accesses by index. eg: list[0]
- dictionary accesses by key(name). eg: dict["age"]
