5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

```python
my_list = [10,20,30]
my_tuple = (10,20,30)

print("Original List:", my_list)
print("Orginal Tuple:", my_tuple)

# list can be modied as is mutable
my_list[1]=200
print("Modified list:", my_list)

# tuple can not be modified and will cause error
try:
  my_tuple[1]=200
except TypeError as e:
  print("Error when modifying tuple:", e)
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def change_values(num, lst):
    num = num + 10        # integers are immutable
    lst.append(10)        # lists are mutable
    return num, lst

x = 5
y = [1, 2, 3]

new_num, new_list = change_values(x, y)

print("Original number:", x)
print("Modified number:", new_num)

print("Original list:", y)
print("Modified list:", new_list)
```

10. Write code to demonstrate type casting with list elements.

```python
num_list = ["10", "20", "30"]

int_list = [int(x) for x in num_list]
float_list = [float(x) for x in num_list]

print("Original list:", num_list)
print("Integer list:", int_list)
print("Float list:", float_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
total = 0

for i in range(1, 51):
    if i % 2 == 0:
        total += i

print("Sum of even numbers:", total)
```

14. Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")

reverse = ""
i = len(text) - 1

while i >= 0:
    reverse += text[i]
    i -= 1

print("Reversed string:", reverse)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

```python
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial:", factorial)
```
